```
firsts = {}
follows = {}
def first(symbol,G):
    global firsts
    if firsts.get(symbol):
        return firsts[symbol]
    else:
        firsts[symbol] = set()
        productions = G["productions"][symbol]
        for p in productions:
            for char in p:
                if char in G["terminals"]:
                    firsts[symbol].add(char)
                    break
                elif char in G["symbols"]:
                    f_char = first(char,G)
                    firsts[symbol] = firsts[symbol].union(f_char)
                    if 'E' not in f_char:
                        break
                    else:
                        if char != p[-1]:
                            firsts[symbol]-=set('E') 
                elif char == 'E':#epsilon
                    if char == p[-1]:
                        firsts[symbol].add('E')
                
                    
    return firsts[symbol]



def follow(symbol,G):
    global follows
    if follows.get(symbol):
        return follows[symbol]
    else:
        follows[symbol] = set()
        if symbol==G["start"]:
            follows[symbol].add('$')
        
        for s,production in G["productions"].items():            
            for p in production:
                ind = p.index(symbol) if symbol in p else -1
                if ind!=-1:
                    if ind+1<len(p):
                        for i in range(ind+1,len(p)):
                            if p[i] in G["terminals"]:
                                follows[symbol].add(p[i])
                                break
                            else:
                                f = first(p[i],G)
                                if 'E' not in f:
                                    follows[symbol] = follows[symbol].union(f)
                                    break
                                else:
                                    if i==len(p)-1:
                                        follows[symbol] = follows[symbol].union(follow(s,G))                                        
                                    follows[symbol] = follows[symbol].union(f-set('E'))
                    else:                            
                        follows[symbol] = follows[symbol].union(follow(s,G))
    return follows[symbol]    

def all_firsts(G):
    global firsts     
    firsts = {}
    for symbol in G["symbols"]:
        first(symbol,G)
        
def all_follows(G):
    global follows
    follows = {}
    for symbol in G["symbols"]:
        follow(symbol,G)
```

```
# epsilon == E
G = {"symbols":set(['S','A','B','C','D','F']), "terminals":set(['a','b','c','d','e']),
     "start":'S', 'productions':{'S':['ABCDF'],'A':['a','E'],'B':['b','E'],'C':['c'],
                                 'D':['d','E'], 'F':['e','E']}}

follow('S',G)
all_firsts(G)
all_follows(G)
print(f"{10*'#'}first{10*'#'}")
print(*[f"{k}: {v}" for k,v in firsts.items()],sep='\n')
print(f"{10*'#'}follow{10*'#'}")
print(*[f"{k}: {v}" for k,v in follows.items()],sep='\n')
```
