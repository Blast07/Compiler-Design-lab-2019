```
import re

def isDelimiter(ch):
    delimiters="- ,;+*/=><!%&|(){}[]]"
    if ch in delimiters:
        return True
    else:
        return False

def isKeyword(string):
    keywords =["int","float","char","double","sizeof","short",
               "while","for","do","break","continue","struct"
               "if","else","elseif","switch","case","typedef"
               "return","void","unsigned","static"]
    if string in keywords:
        return True
    else:
        return False
    
def tokenize(lexeme):
    identifier = r"^[a-zA-Z_][a-zA-Z_0-9]*$"
    const = r"^[0-9][0-9]*$|^[0-9]?.[0-9]+$"
    literal = r'^".*"$'
    operator = r"^[-+/*=><!%&!]$"
    
    if (isKeyword(lexeme)):
        return "keyword"
    elif (re.match(identifier, lexeme)):
        return "identifier"
    elif (re.match(const, lexeme)):
        return "constant"
    elif re.match(literal, lexeme):
        return "literal"
    elif re.match(operator,lexeme):
        return "operator"
    elif isDelimiter(lexeme):
        return "delimiter"
    else:
        return "invalid"
    
    
def parser(string):
    lexemes = []
    start = 0
    i=0
    while i < len(string):
        if isDelimiter(string[i]):
            if (start!=i):
                lexemes.append(string[start:i])
                if(string[i]!=' '):
                    lexemes.append(string[i])
                start = i+1
            else:
                if(string[i]!=' '):
                    lexemes.append(string[i])
                start +=1
        i+=1
    tokens = [(lexeme ,tokenize(lexeme)) for lexeme in lexemes]
    return tokens
string = """int a = myfunc(234,3.14,"multiply")"""   
print(*parser(string),sep='\n')    
```
